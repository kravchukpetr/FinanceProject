```python
from tradernet import NtApi
import os

pub_ = os.environ['FREEDOM_PUBLIC_KEY']
sec_ = os.environ['FREEDOM_SECRET_KEY']

res = NtApi(pub_, sec_, NtApi.V2)
json_position = res.sendRequest('getPositionJson')
```

```python
for index, row in df_pos.iterrows():
    is_found = False
    stock = row['i'].split('.')[0]
    print(stock)
    try:
        ticker = TA_Handler(
            symbol=stock,
            screener="america",
            exchange="NASDAQ",
            interval=Interval.INTERVAL_1_DAY
        )
        print("INTERVAL_1_DAY", ticker.get_analysis().summary)
        ticker = TA_Handler(
            symbol=stock,
            screener="america",
            exchange="NASDAQ",
            interval=Interval.INTERVAL_1_WEEK
        )
        print("INTERVAL_1_WEEK", ticker.get_analysis().summary)
        ticker = TA_Handler(
            symbol=stock,
            screener="america",
            exchange="NASDAQ",
            interval=Interval.INTERVAL_1_MONTH
        )
        print("INTERVAL_1_MONTH", ticker.get_analysis().summary)
        is_found = True
    except:
        pass
    if not is_found:
        try:
            ticker = TA_Handler(
                symbol=stock,
                screener="america",
                exchange="NYSE",
                interval=Interval.INTERVAL_1_DAY
            )
            print("INTERVAL_1_DAY", ticker.get_analysis().summary)
            ticker = TA_Handler(
                symbol=stock,
                screener="america",
                exchange="NYSE",
                interval=Interval.INTERVAL_1_WEEK
            )
            print("INTERVAL_1_DAY", ticker.get_analysis().summary)
            ticker = TA_Handler(
                symbol=stock,
                screener="america",
                exchange="NYSE",
                interval=Interval.INTERVAL_1_MONTH
            )
            print("INTERVAL_1_MONTH", ticker.get_analysis().summary)
        except Exception:
            print("Not parsed")
    
```
